```python
import json
```

```python
def read_json(filename):
    with open(filename, 'r', encoding="utf8") as f:
        return json.load(f)
```

```python
test_dbpedia = read_json('../data/qald_9_plus_test_dbpedia.json')
train_dbpedia = read_json('../data/qald_9_plus_train_dbpedia.json')
test_wikidata = read_json('../data/qald_9_plus_test_wikidata.json')
train_wikidata = read_json('../data/qald_9_plus_train_wikidata.json')
```

```python
def preview_question_list(q):
    question_en = next(filter((lambda x: x.get('language') == 'en'), q.get('question')))
    question_es = next(filter((lambda x: x.get('language') == 'es'), q.get('question')))
    return { 'id': q.get('id'), 'question_en': question_en, 'question_es' : question_es }
```

```python
list(map(preview_question_list, test_dbpedia.get('questions')))
```
