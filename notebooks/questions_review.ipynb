```python
import json
```

```python
def read_json(filename):
    with open(filename, 'r', encoding="utf8") as f:
        return json.load(f)
```

```python
test_dbpedia = read_json('../data/qald_9_plus_test_dbpedia.json')
train_dbpedia = read_json('../data/qald_9_plus_train_dbpedia.json')
test_wikidata = read_json('../data/qald_9_plus_test_wikidata.json')
train_wikidata = read_json('../data/qald_9_plus_train_wikidata.json')
```

```python
def preview_question_list(q):
    question_en = next(filter((lambda x: x.get('language') == 'en'), q.get('question'))).get('string')
    question_es = next(filter((lambda x: x.get('language') == 'es'), q.get('question'))).get('string')
    return { 'id': q.get('id'), 'question_en': question_en, 'question_es' : question_es }
```

```python
test_dbpedia = list(map(preview_question_list, test_dbpedia.get('questions')))
```

```python
test_wikidata = list(map(preview_question_list, test_wikidata.get('questions')))
```

```python
test_questions = test_dbpedia + test_wikidata
```

```python
unique_test_questions = []
for question in test_questions:
    unique_ids = [x.get('id') for x in unique_test_questions]
    if question.get('id') not in unique_ids:
        unique_test_questions.append(question)
```
